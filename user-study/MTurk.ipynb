```python
import os
import glob as glob
import itertools
import pandas as pd
import numpy as np
import json
from scipy.stats import wilcoxon, norm
import math
```

```python
folder = 'UserC1'
```

```python
def getFilename(path):
    return "https://xxxxxxxxx/" + os.path.split(path)[1]
```

```python
humanC = [getFilename(x) for x in glob.glob(os.path.join(folder, "human", "*.png"))]
```

```python
excelC = [getFilename(x) for x in glob.glob(os.path.join(folder, "excelFig", "*.png"))]
```

```python
optC = [getFilename(x) for x in glob.glob(os.path.join(folder, "ours", "*.png"))]
```

```python
randomC = [getFilename(x) for x in glob.glob(os.path.join(folder, "random", "*.png"))]
```

```python
humanC.sort()
excelC.sort()
optC.sort()
randomC.sort()
```

```python
fourwayZip = list(zip(humanC, excelC, optC, randomC))
pairwiseCompare = [list(itertools.combinations(cc, 2)) for cc in fourwayZip]
```

```python
pairs = list(itertools.chain(*pairwiseCompare))
```

```python
np.random.shuffle(pairs)
```

```python
def zipForCompare(pairs, perBatch = 9):
    if len(pairs) % perBatch != 0:
        remaining = perBatch - len(pairs) % perBatch
        pairs.extend(pairs[:remaining])
    
    zipPairs = [list(itertools.chain(*pairs[i:i+perBatch])) for i in range(0, len(pairs), perBatch)]
    columnNames = []
    for i in range(1, perBatch + 1):
        columnNames.append('img' + str(i) + '_1')
        columnNames.append('img' + str(i) + '_2')
    return pd.DataFrame(zipPairs, columns=columnNames)
```

```python
zipForCompare(pairs)
```
